```
# Creator - Rohit Gupta

import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive')

df5 = pd.read_csv('/content/drive/MyDrive/COLAB/EOR.csv')
pd.set_option('display.max_columns', None)

print("EOR Screening Criteria: ")
print(df5.replace(np.nan, "-"))

df = df5.drop([4,6,8])

df.reset_index(drop=True, inplace=True)

col = df.columns.tolist()
l1 = len(col)

col

df2 = pd.DataFrame(data=df.Property, columns=['Property'])

for i in range(1,l1):
  low=[]
  high=[]
  for j in range(0,6):
    s = df[col[i]][j]
    if pd.isna(df[col[i]][j])==True:
      L = 0
      H = 10e10
    elif s[0]=='<':
      L = 0
      H = float(s[1:])
    elif s[0]=='>':
      L = float(s[1:])
      H = 10e10
    else:
      L = s.split('-')[0]
      H = s.split('-')[1]
    low.append(float(L))
    high.append(float(H))
  df2['{}(L)'.format(col[i])] = low
  df2['{}(H)'.format(col[i])] = high

df3 = df2.set_index('Property').transpose()

df3

columns = df3.columns.tolist()

l,p = df3.shape

l2 = len(columns)

columns

l2

def EOR(val):
  k=0
  t=[]
  for j in range(l2):
    c=1
    for i in range(0, l-1, 2):
      L = df3[columns[j]][i]
      H = df3[columns[j]][i+1]
      if L < val[k] and val[k] < H:
        t.append(col[c])
      c=c+1
    k=k+1
  return t

df4 = pd.DataFrame(data=col[1:], columns=['EOR Technique'])

val=[]
print("Please input reservoir parameters to predict best suited EOR Techniques")
val.append(float(input("Oil API: ")))
val.append(float(input("Oil viscosity, cp: ")))
val.append(float(input("Reservoir Depth, ft: ")))
val.append(float(input("Permeability, md: ")))
val.append(float(input("Oil saturation, %: ")))
val.append(float(input("Salinity, ppm: ")))

len(val)

EORtech = EOR(val)

EORtech

c=[]
for i in range(1, l1):
  Count = EORtech.count(col[i])
  c.append(Count)
df4['Count'] = c

df4

max = df4['Count'].max()
EORtech = []
df4 = df4[df4['Count']==max]
lis = df4['EOR Technique'].tolist()

s=''
for items in lis:
  s = s+items+', '
s = s[:len(s)-2]

if len(lis)==0:
  print("Input parameter/s are imaginary")
else:
  print('EOR method/s to be used is/are {} flooding.'.format(s))
```
